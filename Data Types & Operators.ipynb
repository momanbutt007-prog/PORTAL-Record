```python
# ===============================
# Python Data Types & Operators
# ===============================

# Variables (Different Data Types)
name = "Momin Butt"
age = 24
height = 5.8
is_student = True

# List, Tuple, Dictionary
subjects = ["Politics", "Physics", "CS"]
marks = (85, 90, 88)
student = {
    "name": name,
    "age": age,
    "city": "Gujranwala"
}

# ===============================
# Arithmetic Operators
# ===============================
num1 = int (input('enter first number'))
num2 =int(input('Enter second number'))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)
print("-------------------------------------------------")

# ===============================
# Comparison Operators
# ===============================
print("Is num1 greater than num2?", num1 > num2)
print("Are both equal?", num1 == num2)
print("-------------------------------------------------")
# ===============================
# Logical Operators
# ===============================
print("Logical AND:", num1 > 5 and num2 < 10)
print("Logical OR:", num1 < 5 or num2 < 10)
print("Logical NOT:", not(is_student))
print("-------------------------------------------------")
# ===============================
# Assignment Operators
# ===============================
num1 += 5
print("After += :", num1)

num2 -= 2
print("After -= :", num2)
print("-------------------------------------------------")
# ===============================
# Membership Operators
# ===============================
print("Is 'Math' in subjects?", "Math" in subjects)
print("Is 'Biology' not in subjects?", "Biology" not in subjects)
print("-------------------------------------------------")
# ===============================
# Output Student Info
# ===============================
print("\n--- Student Info ---")
print("Name:", student["name"])
print("Age:", student["age"])
print("City:", student["city"])
print("Subjects:", subjects)
print("Marks:", marks)
```
