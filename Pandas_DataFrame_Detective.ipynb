```python
import pandas as pd
import numpy as np
```

```python

# 2. Create dictionary
data = {
    'Name': ['Alice','Bob','Charlie','Diana','Eve','Frank'],
    'Age': [25,30,np.nan,35,28,np.nan],
    'Score': [85,92,78,np.nan,88,95],
    'Grade': ['A','A','B','C','B','A']
}
print("------------------------------------------")

# 3. Create DataFrame
df = pd.DataFrame(data)

# 4. Basic Info
print("Head (first 3 rows):\n", df.head(3))
print("\nShape:", df.shape)
print("\nColumns:", df.columns)
print("\nData Types:\n", df.dtypes)
print("------------------------------------------")

# 5. Missing Data
print("\nMissing Values:\n", df.isnull())
print("------------------------------------------")

# Fill Age with mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
print("------------------------------------------")

# Drop rows with missing Score
df = df.dropna(subset=['Score'])

print("\nAfter Handling Missing Data:\n", df)
print("------------------------------------------")

# 6. Selection
print("\nName and Score:\n", df[['Name','Score']])

print("\nRows where Score > 85:\n", df[df['Score'] > 85])

print("------------------------------------------")
# Set Name as index
df.set_index('Name', inplace=True)
print("\nDataFrame with Name as Index:\n", df)
print("------------------------------------------")
# 7. Operations
print("\nUnique Grades:", df['Grade'].unique())

print("\nGrade Counts:\n", df['Grade'].value_counts())

print("\nSorted by Score (Descending):\n", df.sort_values(by='Score', ascending=False))
```
