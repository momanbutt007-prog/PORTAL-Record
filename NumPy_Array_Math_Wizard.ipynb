```python
import numpy as np 
```

```python
# 2. Create arrays
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([10,20,30])
# 3. Operations using broadcasting
print("Addition:\n", arr1 + arr2)
print("---------------------------------------")
print("Subtraction:\n", arr1 - arr2)
print("---------------------------------------")
print("Multiplication:\n", arr1 * arr2)
print("---------------------------------------")
print("Division:\n", arr1 / arr2)

print("---------------------------------------")
# 4. Broadcasting Explanation
# Broadcasting allows NumPy to perform operations on arrays of different shapes.
# Here, arr2 (1D) is automatically expanded to match arr1 (2D) shape.

# 5. Universal Functions
arr3 = np.array([1,4,9,16,25])

print("\nSquare Root:", np.sqrt(arr3))
print("Exponential:", np.exp(arr3))
print("Sine:", np.sin(arr3))
print("---------------------------------------")

# 6. Array operations
arr4 = np.array([[5,12,3],[8,2,7]])

print("\nSum of all elements:", np.sum(arr4))
print("Max in each column:", np.max(arr4, axis=0))
print("Min in each row:", np.min(arr4, axis=1))
print("---------------------------------------")

# 7. Type casting
arr1_float = arr1.astype(float)
print("\narr1 as float:\n", arr1_float)
print("---------------------------------------")
```
